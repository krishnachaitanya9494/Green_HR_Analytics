```python
#income vs Attrition


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="Attrition", y="MonthlyIncome", palette="Set3")
plt.title("Monthly Income vs Attrition")
plt.show()
plt.tight_layout()
```

```python
income_mean = df.groupby("Attrition")["MonthlyIncome"].mean()
print(income_mean)
```

```python
#satisfaction vs worklife

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.countplot(data=df, x="WorkLifeBalance", hue="Attrition", palette="Set3")
plt.title("job satisfaction vs Attrition")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(data=df, x="Age", hue="Attrition", kde=True)
plt.title("Attrition by Age")

plt.subplot(1,2,2)
sns.histplot(data=df, x="YearsAtCompany", hue="Attrition", kde=True)
plt.title("Attrition by Tenure")

plt.tight_layout()
plt.show()
attrition_age = df.groupby("Attrition")["Age"].mean()
print(attrition_age)
attrition_tenure = df.groupby("Attrition")["YearsAtCompany"].mean()
print(attrition_tenure)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
dept_attr = df.groupby("Department")["Attrition"].value_counts(normalize=True).mul(100).rename("Rate").reset_index()

sns.barplot(data=dept_attr[dept_attr["Attrition"]=="Yes"],
            x="Department", y="Rate", palette="viridis")
plt.title("Attrition Rate by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
#salary by department

df.groupby('Department')['MonthlyIncome'].mean()


#Salary by job role
df.groupby('JobRole')['MonthlyIncome'].mean()


```

```python
# Overall distributions
df['JobSatisfaction'].value_counts()
df['EnvironmentSatisfaction'].value_counts()
df['RelationshipSatisfaction'].value_counts()

# Compare satisfaction between attrition and non-attrition
df.groupby('Attrition')[['JobSatisfaction',
                         'EnvironmentSatisfaction',
                         'RelationshipSatisfaction']].mean()
```
